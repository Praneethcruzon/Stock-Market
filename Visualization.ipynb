```
import pandas as pd
from sklearn import preprocessing
data = pd.read_csv("BMW.DE.csv", sep = ',')
data = data.dropna(axis = 0, how ='any') 
```

```
open = data.rolling(window = 600)['Open'].mean()
open = pd.DataFrame(open)
volume = data.rolling(window = 600)['Volume'].mean()
volume = pd.DataFrame(volume)
min_max_scaler = preprocessing.MinMaxScaler()
openscaled = min_max_scaler.fit_transform(open)
volumescaled = min_max_scaler.fit_transform(volume)
```

```
from matplotlib import pyplot as plt
plt.figure()
plt.plot(openscaled)
plt.plot(volumescaled)
plt.show()
```
